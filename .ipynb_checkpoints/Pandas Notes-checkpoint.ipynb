# TASK 3: Print a list of all establishment names that are missing any item of information

# TASK 5: A count of how many establishments received each rating: 1, 2, 3, 4 and 5, and how many ratings were omitted

```python
import csv

with open('foodhygienedata.csv','r') as file:
    reader = csv.DictReader(file,delimiter = ',')
    rating1 = 0
    rating2 = 0
    rating3 = 0
    rating4 = 0
    rating5 = 0
    missing = 0
    for row in reader:
        if row['rating'] =='':
            missing += 1
        elif int(row['rating']) == 1:
            rating1 += 1
        elif int(row['rating']) == 2:
            rating2 += 1
        elif int(row['rating']) == 3:
            rating3 += 1
        elif int(row['rating']) == 4:
            rating4 += 1
        elif int(row['rating']) == 5:
            rating5 += 1

            
    print('Amount scoring 1: {0}'.format(rating1))
    print('Amount scoring 2: {0}'.format(rating2))
    print('Amount scoring 3: {0}'.format(rating3))
    print('Amount scoring 4: {0}'.format(rating4))
    print('Amount scoring 5: {0}'.format(rating5))
    print('Amount missing: {0}'.format(missing))
    
            
```

# Creating, writing and appending new file

```python
import csv

with open('modules.csv','r') as readFile:
    reader = csv.DictReader(readFile,delimiter = ',')

    fieldnames = reader.fieldnames
    print(fieldnames)
    
    with open('newFile.csv','w') as writeFile:
        writer = csv.DictWriter(writeFile, fieldnames = fieldnames, delimiter =',')
        
        for row in reader:
          writer.writerow(row)
        
    newLine = {'Module Name': 'HTML CSS JS', ' Module Code': 'COM999'}
    
    with open('newFile.csv','a') as writeFile:
        appender = csv.DictWriter(writeFile, fieldnames = fieldnames, delimiter = ',')
        appender.writerow(newLine)
        
```

# TASK 6: Read csv data from Food Hygiene Dataset. Work on data and write to a new file

```python
import csv
from datetime import date

with open('foodhygienedata.csv','r') as file:
    reader = csv.DictReader(file, delimiter = ',')
    
#   fieldnames = reader.fieldnames
    fieldnames = ['establishmentname','inspectiondate','dayssinceinspection']
    
    with open('daysSinceInspection.csv','w') as newFile:
        writer = csv.DictWriter(newFile, fieldnames = fieldnames, delimiter = ',')
        writer.writeheader()
        
        newDict = {}
        
        for row in reader:
            inspectiondate = row['inspectiondate']
            if not (row['inspectiondate']):
                continue
            else:
                day, month, year = inspectiondate.split('/')
                d0 = date(int(year), int(month), int(day))
                d1 = date.today()
                delta = d1 - d0
                print("{0} was inspected on {1}. {2} days since inspection".format(row['establishmentname'], inspectiondate, delta.days))
                newDict['establishmentname'] = row['establishmentname']
                newDict['inspectiondate'] = inspectiondate
                newDict['dayssinceinspection'] = delta.days
                writer.writerow(newDict)
```

# Compare speed of Numpy with Python array

```python
import numpy as np
import time
import sys

size = 1000000

#Python list
l1 = range(size)
l2 = range(size)

#Numpy array
a1 = np.arange(size)
a2 = np.arange(size)

#Python list
start = time.time()
result = [(x+y) for x,y in zip(l1,l2)]
print("Python list took: ",(time.time() - start)*1000)

#Numpy array
start = time.time()
result = a1 + a2
print('Numpy list took: ',(time.time()-start)*1000)
```

# Numpy Tutorial

```python
import numpy as np

a = np.array([1,2,3])
print("a = ",a)

b = np.arange(1,13,2)
print("b = ",b)

c = np.linspace(1,13,6)
print("c = ",c)

d = c.reshape(3,2)
print("d = ",d)
print('d size: ',d.size)
print('d shape: ',d.shape)
print('d data type: ',d.dtype)
print('d item size: ',d.itemsize, " bytes each")

e = np.array([(1,2,3),(4,5,6)])
print("e = ",e)

f = e < 4
print("f = ",f)

g = e * 4
print("g = ",g)

h = np.array([1,2,3])
h *= 2
print("h = ",h)

i = np.zeros((3,4))
print("i = ",i)

j = np.ones((2,3))
print("j = ",j)

k = np.ones(10)
print("k = ",k)
print("k dtype = ",k.dtype)

l = np.array([1,2,3], dtype = np.int16)
print("l array with dtype np.int16 = ",l)

m = np.random.random((2,3))
print(m)
print("m dtype = ",m.dtype)
print("m itemsize = ",m.itemsize," bytes each")

np.set_printoptions(precision = 2, suppress = True)
print("np.set_printoptions(precision = 2, suppress = True)")

n = np.random.randint(0,11,20)
print("n = ",n)
print("n sum = ",n.sum())
print("n min = ",n.min())
print("n max = ",n.max())
print("n mean = ",n.mean())
print("n var = ",n.var())
print("n std = ",n.std())

o = np.random.randint(0,11,6)
print("o = ",o)

o = o.reshape(3,2)
print("o reshape = ",o)
print("o sum axis 1 = ",o.sum(axis = 1))
print("o sum axis 0 = ",o.sum(axis = 0))

p = np.arange(0,11)
print("p = ",p)

np.random.shuffle(p)
print("p random shuffle = ",p)

q = np.arange(0,11)
print("q = ",q)
print("q random choice = ",np.random.choice(q))

r = np.random.randint(1,11,2)
print("r random randint = ",r)
```

```python
import numpy as np
import pandas as pd

def header(msg):
    print('-' * 50)
    print('[' + msg + ']')
    print('-' * 50)
    
df = pd.DataFrame(
    [
        ['Jan',12,12,34.5],
        ['Feb',11,42,34.5],
        ['Mar',14,22,64.5],
        ['Apr',15,42,74.5],
    ],
    index = [0,1,2,3],
    columns = ['Month', 'Day Temp',' Night Temp', 'Average']
)

header('1. Load hard-coaded data into a dataframe')
print(df)
```
